```python
sentence = correct_samples[3]
```

```python
# sentence = 'Says Hillary Clinton has donated every cent shes ever earned from speaking fees to charity.'
# sentence = 'Says President Barack Obama spent nearly a trillion dollars on make-work government jobs.'
sentence = 'Says Bill Clinton opposes President Barack Obama plan to raise taxes on wealthy Americans.'
```

```python
from bertviz import head_view, model_view
from transformers import RobertaTokenizer, RobertaModel
from pathlib import Path

tokenizer = RobertaTokenizer.from_pretrained(Path(str(model_dir)), use_fast=True)
model = RobertaModel.from_pretrained(
    str(model_dir),
    output_attentions=True
)

inputs = tokenizer.encode_plus(sentence, return_tensors='pt')
input_ids = inputs['input_ids']
attention = model(input_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
```

```python
head_view(attention, tokens)
```

```python
model_view(attention, tokens)
```
